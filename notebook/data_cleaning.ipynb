## Step 4: CLEAN MATCHES TABLE

- HANDLE MISSING VALUES

```python
matches.fillna({'toss_decision': "Unknown"}, inplace=True)
matches.fillna({'toss_winner': "Unknown"}, inplace=True)
matches.fillna({'match_winner': "No Result"}, inplace=True)
```

- STANDARDIZE TEAM NAMES

```python
def clean_team_name(name):
    if pd.isna(name):
        return name
    name = name.strip().lower()
    
    mapping = {
        'mumbai indians': 'MI',
        'chennai super kings': 'CSK',
        'royal challengers bangalore': 'RCB',
        'kolkata knight riders': 'KKR',
        'delhi capitals': 'DC',
        'sunrisers hyderabad': 'SRH',
        'rajasthan royals': 'RR',
        'punjab kings': 'PBKS',
        'Gujarat Titans': 'GT',
        'Lucknow Super Giants': 'LSG',
        'gujarat titans': 'GT',
        'lucknow super giants': 'LSG',
        'deccan chargers': 'SRH',  # better mapping than DC
        'rising pune supergiant': 'RPS',
        'rising pune supergiants': 'RPS',
        'gujarat lions': 'GL',
        'pune warriors': 'PWI',
        'kochi tuskers kerala': 'KTK',
        'delhi daredevils': 'DC'
    }
    
    return mapping.get(name, name.upper())

matches['team1'] = matches['team1'].apply(clean_team_name)
matches['team2'] = matches['team2'].apply(clean_team_name)
matches['match_winner'] = matches['match_winner'].apply(clean_team_name)
matches['toss_winner'] = matches['toss_winner'].apply(clean_team_name)
```

- CONVERT DATE PROPERLY

```python
matches['match_date'] = pd.to_datetime(matches['match_date'], errors='coerce')
```

- REMOVE DUPLICATES

```python
matches.drop_duplicates(inplace=True)
```

## Step 5: CLEAN DELIVERIES TABLE

```python
deliveries.drop_duplicates(inplace=True)
```

- STANDARDIZE TEAM NAMES

```python
def clean_team_name(name):
    if pd.isna(name):
        return name
    name = name.strip().lower()
    
    mapping = {
        'mumbai indians': 'MI',
        'chennai super kings': 'CSK',
        'royal challengers bangalore': 'RCB',
        'kolkata knight riders': 'KKR',
        'delhi capitals': 'DC',
        'sunrisers hyderabad': 'SRH',
        'rajasthan royals': 'RR',
        'punjab kings': 'PBKS',
        'Gujarat Titans': 'GT',
        'Lucknow Super Giants': 'LSG',
        'gujarat titans': 'GT',
        'lucknow super giants': 'LSG',
        'deccan chargers': 'SRH',  # better mapping than DC
        'rising pune supergiant': 'RPS',
        'rising pune supergiants': 'RPS',
        'gujarat lions': 'GL',
        'pune warriors': 'PWI',
        'kochi tuskers kerala': 'KTK',
        'delhi daredevils': 'DC'
    }
    
    return mapping.get(name, name.upper())

deliveries['team_batting'] = deliveries['team_batting'].apply(clean_team_name)
deliveries['team_bowling'] = deliveries['team_bowling'].apply(clean_team_name)
```
